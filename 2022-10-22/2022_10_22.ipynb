```python
# Data Wrangling Practice

# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset from URL

url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
dataset = pd.read_csv(url, sep = '\t')
```

```python
# Printing the first 10 rows of the dataset

dataset.head(10)
```

```python
# Dropt the order_id column and choice_description column

dataset.drop(['order_id', 'choice_description'], axis = 1, inplace = True)

# Printing the first 10 rows of the dataset

dataset.head(10)
```

```python
# How many products cost more than $10.00?

dataset[dataset.item_price > '$10.00'].item_name.nunique()

# What was the quantity of the most expensive item ordered?

dataset[dataset.item_price == dataset.item_price.max()].quantity

# How many times were a Veggie Salad Bowl ordered?

dataset[dataset.item_name == 'Veggie Salad Bowl'].quantity.sum()

# Print the answer to the question above

print('Veggie Salad Bowl was ordered {} times'.format(dataset[dataset.item_name == 'Veggie Salad Bowl'].quantity.sum()))
print ('The quantity of the most expensive item ordered was {}'.format(dataset[dataset.item_price == dataset.item_price.max()].quantity))
print ('{} products cost more than $10.00'.format(dataset[dataset.item_price > '$10.00'].item_name.nunique()))
```

```python
# Plot the median price of the items

from matplotlib import pyplot as plt
```

```python
# Plot the median price of the items

dataset.groupby('item_name').item_price.median().sort_values(ascending = False).plot(kind = 'bar', figsize = (10, 5)) # This command plots the median price of the items
plt.ylabel('Median Price')
plt.xlabel('Item Name')
plt.title('Median Price of the Items')
plt.show()

```
